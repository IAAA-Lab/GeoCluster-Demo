## Clustering with a combined distance
### Dataset Zgz

This notebook executes CombinedApplication to create a clusters shapefile from the data contained in the specified dataset.

This dataset contains localities in the Spanish area of Zaragoza. Data consists in the polygon geometry and a numeric attribute - the population of a locality.

```python
%matplotlib inline

import subprocess
import matplotlib.pyplot as plt
import shapefile
```

### Execution 1

** Algorithm: ** DBSCAN (eps = 0.5, minPts = 2)

** Distance: ** Hausdorff

** Results: ** 10 clusters are generated instead of the 15 expected. This is because some of the groups are too near, so DBSCAN considers them neighbors and puts them together.

```python
subprocess.call(['java', '-jar', 'combined.jar', 'zgz', '0.6', '0.4', '5000', '1'], stdout=None, stderr=None)

# Plot elements colored by cluster
colors = ['b','g','r','c','m','k','y','#eeefff','#abefab','#eaecab','#effaab']
sf = shapefile.Reader("cluster")
i = 0
for sr in sf.shapeRecords():
    shapePoints = []
    for x,y in sr.shape.points:
        plt.plot(x,y, marker='o',color=colors[i])
    i = i + 1
plt.show()
```

### Execution 2

** Algorithm: ** DBSCAN (eps = 0.5, minPts = 2)

** Distance: ** Euclidean

** Results: ** same results with execution 1, because the only change is the distance (Hausdorff vs Euclidean), and there is no difference between them when the elements are single points.

```python
subprocess.call(['java', '-jar', 'application.jar', 'R15', 'Euclidean', 'DBSCAN', '0.5', '2'], stdout=None, stderr=None)

# Plot elements colored by cluster
colors = ['b','g','r','c','m','k','y','#eeefff','#abefab','#eaecab','#effaab']
sf = shapefile.Reader("cluster")
i = 0
for sr in sf.shapeRecords():
    shapePoints = []
    for x,y in sr.shape.points:
        plt.plot(x,y, marker='o',color=colors[i])
    i = i + 1
plt.show()
```

### Execution 3

** Algorithm: ** DBSCAN (eps = 0.35, minPts = 10)

** Distance: ** Euclidean

** Results: ** we have changed the parameters of the algorithm. By reducing eps we are trying to separate those big clusters that have been created in execution 2. By increasing minPts we are discarding mini-clusters that might be created as a consequence of reducing eps. Now we are getting the desired 15 clusters, but if we take a look we can see that we have lost some elements. They are classified as noise for being separate enough to the corresponding cluster.

```python
subprocess.call(['java', '-jar', 'application.jar', 'R15', 'Euclidean', 'DBSCAN', '0.35', '10'], stdout=None, stderr=None)

# Plot elements colored by cluster
colors = ['b','g','r','c','m','k','y','#eeefff','#abefab','#eaecab','#effaab','#ff9966','#ff66ff','#6699ff','#996633']
sf = shapefile.Reader("cluster")
i = 0
for sr in sf.shapeRecords():
    shapePoints = []
    for x,y in sr.shape.points:
        plt.plot(x,y, marker='o',color=colors[i])
    i = i + 1
plt.show()
```

### Execution 4

** Algorithm: ** KMeans (k = 15)

** Distance: ** Euclidean

** Results: ** 15 clusters are generated because they have been specified by K-Means. Results are different in every execution because of the randomized initialization of centroids. In general, the problems of K-Means can be seen here: if two centroids are put too near, we may get unwanted results.

```python
subprocess.call(['java', '-jar', 'application.jar', 'R15', 'Euclidean', 'KMeans', '15'], stdout=None, stderr=None)

# Plot elements colored by cluster
colors = ['b','g','r','c','m','k','y','#eeefff','#abefab','#eaecab','#effaab','#ff9966','#ff66ff','#6699ff','#996633']
sf = shapefile.Reader("cluster")
i = 0
for sr in sf.shapeRecords():
    shapePoints = []
    for x,y in sr.shape.points:
        plt.plot(x,y, marker='o',color=colors[i])
    i = i + 1
plt.show()
```

### Execution 5

** Algorithm: ** KMeans++ (k = 15)

** Distance: ** Euclidean

** Results: ** again, 15 clusters are generated and results will not be the same in every execution. But in this case we are using K-Means++ so initial centroids will tend to be separated. Even though, we can see bad results. This is due to the elements being so concentrated in the space, so having separated centroids does not guarantee having them in the places where there is a concentration of points.

```python
subprocess.call(['java', '-jar', 'application.jar', 'R15', 'Euclidean', 'KMeans++', '15'], stdout=None, stderr=None)

# Plot elements colored by cluster
colors = ['b','g','r','c','m','k','y','#eeefff','#abefab','#eaecab','#effaab','#ff9966','#ff66ff','#6699ff','#996633']
sf = shapefile.Reader("cluster")
i = 0
for sr in sf.shapeRecords():
    shapePoints = []
    for x,y in sr.shape.points:
        plt.plot(x,y, marker='o',color=colors[i])
    i = i + 1
plt.show()
```

### Conclusions
This dataset consists in a space with high-density areas in the middle of a big, empty space. That's why DBSCAN algorithm works best. Also, due to the data being single points, Hausdorff and Euclidean are equal, so Euclidean is enough for being the most simple.
